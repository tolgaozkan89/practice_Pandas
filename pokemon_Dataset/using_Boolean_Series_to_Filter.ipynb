# Create some Filters with Boolean Series

## Attack more then 75 (more_attack)

```python
more_attack = pokemon["Attack"] >= 75
```

```python
more_attack
```

```python
more_attack.shape
```

## Speed more then 75 (more_speed)

```python
more_speed = pokemon["Speed"] >= 75
```

```python
more_speed.shape
```

# Question 1: 

# How many pokemons has more attack than 75 and more speed then 75?

```python
pokemon_more_attack_speed = pokemon[more_attack & more_speed]
```

```python
pokemon_more_attack_speed
```

```python
pokemon_more_attack_speed.shape
```
